```python
import pandas as pd

df = pd.read_csv('../data/bn_translations_2000_2025.csv', dtype={'year': str}, low_memory=False)
print(f"Total records: {len(df)}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nSample:")
df.head()
```

```python
# top languages overall
top_langs = df['lang_original'].value_counts().head(15)
print(top_langs)
```

```python
# Ukrainian vs Russian translations over time
pivot = df[df['lang_original'].isin(['ukr', 'rus'])].groupby(['year', 'lang_original']).size().unstack(fill_value=0)
print(pivot)
```

```python
# shares as percentage of total translations per year
yearly_total = df[df['year'].str.match(r'^\d{4}$')].groupby('year').size()
ukr_rus = df[df['lang_original'].isin(['ukr', 'rus'])].groupby(['year', 'lang_original']).size().unstack(fill_value=0)
ukr_rus_pct = ukr_rus.div(yearly_total, axis=0) * 100
print(ukr_rus_pct.round(2))
```

```python
# top 20 authors translated from Ukrainian and Russian
for lang in ['ukr', 'rus']:
    print(f"\nTop 20 authors translated from {lang}:")
    top = df[df['lang_original'] == lang]['author'].value_counts().head(20)
    print(top)
```

```python
# Russian authors - filter out pre-1900 by birth year
modern_rus = df[
    (df['lang_original'] == 'rus') & 
    (~df['author'].str.contains(r'18\d\d', na=False))
]
print(f"Total Russian: {len(df[df['lang_original'] == 'rus'])}")
print(f"Modern Russian (born after 1900): {len(modern_rus)}")
```

```python
# compare modern authors only
modern_ukr = df[
    (df['lang_original'] == 'ukr') & 
    (~df['author'].str.contains(r'18\d\d', na=False))
]
print(f"Total Ukrainian: {len(df[df['lang_original'] == 'ukr'])}")
print(f"Modern Ukrainian (born after 1900): {len(modern_ukr)}")
print(f"\nModern Russian vs Modern Ukrainian ratio: {len(modern_rus)/len(modern_ukr):.1f}x")
```

```python
# modern authors ratio by year
modern = df[~df['author'].str.contains(r'18\d\d', na=False)]
modern_yearly = modern[modern['lang_original'].isin(['ukr', 'rus'])].groupby(['year', 'lang_original']).size().unstack(fill_value=0)
print(modern_yearly[modern_yearly.index.str.match(r'^\d{4}$')].tail(10))
```

```python
# Gogol translations over time
gogol = df[df['author'].str.contains('Gogol', na=False, case=False)]
print(f"Total Gogol translations: {len(gogol)}")
print(f"\nBy year:")
print(gogol.groupby('year').size().sort_index())
print(f"\nTitles:")
print(gogol['title'].value_counts().head(20))
```

```python
# key events timeline — total translations per year + top languages
yearly = df[df['year'].str.match(r'^\d{4}$')].groupby('year').size()
top5 = df[df['lang_original'].isin(['eng', 'ger', 'fre', 'rus', 'ukr'])]
top5_yearly = top5.groupby(['year', 'lang_original']).size().unstack(fill_value=0)

print("Total translations per year:")
print(yearly[yearly.index >= '2000'])
print("\nTop 5 languages by year:")
print(top5_yearly[top5_yearly.index >= '2000'])
```

```python
# neighbouring countries languages
neighbours = {
    'ger': 'Germany/Austria',
    'cze': 'Czech Republic',
    'slo': 'Slovakia', 
    'ukr': 'Ukraine',
    'bel': 'Belarus',
    'rus': 'Russia/Belarus',
    'lit': 'Lithuania'
}

neighbour_langs = ['ger', 'cze', 'slo', 'ukr', 'bel', 'rus', 'lit']
neigh_data = df[df['lang_original'].isin(neighbour_langs)].groupby(['year', 'lang_original']).size().unstack(fill_value=0)
print(neigh_data[neigh_data.index.str.match(r'^\d{4}$')])
```

```python
# top 20 authors translated from English
print("Top 20 authors translated from English:")
print(df[df['lang_original'] == 'eng']['author'].value_counts().head(20))
```

```python
# top publishers overall
print("Top 20 publishers:")
print(df['publisher'].value_counts().head(20))
```

```python
# who publishes Ukrainian translations
ukr_publishers = df[df['lang_original'] == 'ukr']['publisher'].value_counts().head(20)
print("Top publishers of Ukrainian translations:")
print(ukr_publishers)
```

```python
rus_publishers = df[df['lang_original'] == 'rus']['publisher'].value_counts().head(10)
print("Top publishers of Russian translations:")
print(rus_publishers)
```

```python
# who started publishing Ukrainian after 2022
pre = df[(df['lang_original'] == 'ukr') & (df['year'] < '2022')]['publisher'].value_counts().head(10)
post = df[(df['lang_original'] == 'ukr') & (df['year'] >= '2022')]['publisher'].value_counts().head(10)
print("Ukrainian publishers BEFORE 2022:")
print(pre)
print("\nUkrainian publishers AFTER 2022:")
print(post)
```
